```python
%%writefile palindrome.c
#include <stdio.h>
int main()
{
  int num, remainder, orginal, reverse;
  printf("Enter the number: ") ;
  scanf("%d", &num);
  orginal = num;
  while (num != 0)
  {
    remainder = num % 10;
    reverse = reverse * 10 + remainder;
    num /= 10;
  }
  if (orginal == reverse)
    printf("%d is a palindrome", orginal);
  else
  {
    printf("%d is not a palindrome", orginal);
  }
  return 0;
}
```
